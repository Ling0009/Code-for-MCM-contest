```python
from numpy import *
df2_new=df2.copy()
df2_new['变异系数']=df2.apply(lambda x:x.std(),axis =1)/df2.apply(lambda x:x.mean(),axis =1)
df2_new['变异系数']=df2_new['变异系数']/sqrt(240)
# annualVolatility = std(logreturns) / mean(logreturns)
# annualVolatility = annualVolatility / sqrt(1/252)
df2_new["总和"] =df2.apply(lambda x:x.sum(),axis =1)
df2_new["均值"] =df2.apply(lambda x:x.mean(),axis =1)
df2_new["标准差"] =df2.apply(lambda x:x.mean(),axis =1)
df2_new['零值率']=(df2 == 0).sum(axis=1)
df2_new
```

```python
def regularit(df2_new02):
    df2_new03 = pd.DataFrame(index=df2_new02.index)
    columns = df2_new02.columns.tolist()
    for c in columns:
        d = df2_new02[c]
        MAX = d.max()
        MIN = d.min()
        df2_new03[c] = ((d - MIN) / (MAX - MIN)).tolist()
    return df2_new03
df2_new03=regularit(df2_new02)
#df2_new03=df2_new03.T
df2_new03
```
